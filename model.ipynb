```python

corr_matrix = df.iloc[:, :-1].corr().abs()  # exclude 'target' column
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
to_drop = [column for column in upper.columns if any(upper[column] > 0.8)]
df.drop(columns=to_drop, inplace=True)

print(f"✅ Dropped {len(to_drop)} highly correlated features")
```

```python
from sklearn.model_selection import train_test_split

# Separate features and target again after dropping columns
X = df.drop("target", axis=1)
y = df["target"]

# Split the data (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

print("✅ Data split complete!")
print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)
```

```python
from sklearn.naive_bayes import GaussianNB

# Initialize and train the model
nb_model = GaussianNB()
nb_model.fit(X_train, y_train)

print("✅ Model training complete!")
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# Make predictions
y_pred = nb_model.predict(X_test)

# Evaluate the model
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("Accuracy Score:", accuracy_score(y_test, y_pred))
```

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import RandomizedSearchCV
from sklearn.base import clone
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

# Classifiers
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier

def classification_model_finetuning_rscv(X_train, y_train, X_test, y_test, preprocessor):
    models = {
        "Logistic Regression": LogisticRegression(max_iter=1000),
        "Random Forest": RandomForestClassifier(random_state=42),
        "Gradient Boosting": GradientBoostingClassifier(random_state=42),
        "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42),
        "LightGBM": LGBMClassifier(random_state=42),
        "Naive Bayes": GaussianNB(),
        "SVC": SVC(probability=True)
    }

    param_grid = {
        "Logistic Regression": {'classifier__C': [0.01, 0.1, 1, 10]},
        "Random Forest": {
            'classifier__n_estimators': [100, 200],
            'classifier__max_depth': [None, 10, 20]
        },
        "Gradient Boosting": {
            'classifier__n_estimators': [100, 200],
            'classifier__learning_rate': [0.01, 0.1, 0.2],
            'classifier__max_depth': [3, 5, 7]
        },
        "XGBoost": {
            'classifier__n_estimators': [100, 200],
            'classifier__learning_rate': [0.01, 0.1, 0.2],
            'classifier__max_depth': [3, 5, 7]
        },
        "LightGBM": {
            'classifier__n_estimators': [100, 200],
            'classifier__learning_rate': [0.01, 0.1, 0.2],
            'classifier__num_leaves': [31, 50, 70]
        },
        "SVC": {
            'classifier__C': [0.1, 1, 10],
            'classifier__kernel': ['linear', 'rbf']
        }
    }

    base_scores = {}
    tuned_scores = {}
    best_models = {}

    for name, model in models.items():
        print(f"\n🔧 Training & Evaluating: {name}")

        base_pipeline = Pipeline([
            ("preprocessor", preprocessor),
            ("classifier", model)
        ])

        # Base model training
        base_pipeline.fit(X_train, y_train)
        y_pred = base_pipeline.predict(X_test)
        y_proba = base_pipeline.predict_proba(X_test)[:, 1] if hasattr(base_pipeline, "predict_proba") else y_pred

        base_scores[name] = {
            "Accuracy": accuracy_score(y_test, y_pred),
            "Precision": precision_score(y_test, y_pred),
            "Recall": recall_score(y_test, y_pred),
            "F1": f1_score(y_test, y_pred),
            "ROC AUC": roc_auc_score(y_test, y_proba)
        }

        # Tuning if applicable
        if name in param_grid:
            print(f"→ Running RandomizedSearchCV for {name}")
            try:
                tuned_pipeline = Pipeline([
                    ("preprocessor", preprocessor),
                    ("classifier", clone(model))
                ])

                grid = RandomizedSearchCV(
                    estimator=tuned_pipeline,
                    param_distributions=param_grid[name],
                    n_iter=10,
                    scoring="f1",
                    cv=5,
                    random_state=42,
                    n_jobs=-1
                )
                grid.fit(X_train, y_train)
                best_model = grid.best_estimator_
                tuned_pred = best_model.predict(X_test)
                tuned_proba = best_model.predict_proba(X_test)[:, 1] if hasattr(best_model, "predict_proba") else tuned_pred

                tuned_scores[name] = {
                    "Accuracy": accuracy_score(y_test, tuned_pred),
                    "Precision": precision_score(y_test, tuned_pred),
                    "Recall": recall_score(y_test, tuned_pred),
                    "F1": f1_score(y_test, tuned_pred),
                    "ROC AUC": roc_auc_score(y_test, tuned_proba)
                }

                best_models[name] = best_model
                print("✅ Best Params:", grid.best_params_)
            except Exception as e:
                print(f"⚠ Error tuning {name}: {e}")
                tuned_scores[name] = base_scores[name]
                best_models[name] = base_pipeline
        else:
            tuned_scores[name] = base_scores[name]
            best_models[name] = base_pipeline

    return base_scores, tuned_scores, best_models
```

```python
from sklearn.preprocessing import StandardScaler

# Use previously split data: X_train, X_test, y_train, y_test
preprocessor = StandardScaler()

# Run training
print("=== Starting Classification Model Training with RandomizedSearchCV ===")
base_scores, tuned_scores, best_models = classification_model_finetuning_rscv(
    X_train, y_train, X_test, y_test, preprocessor
)
```

```python
print("\n=== Classification Model Performance Summary ===")
print(f"{'Model':<20} {'Base Acc':<10} {'Tuned Acc':<10} {'Base F1':<10} {'Tuned F1':<10}")
print("-" * 60)

for model_name in base_scores.keys():
    base_acc = base_scores[model_name]['Accuracy']
    base_f1 = base_scores[model_name]['F1']
    
    tuned_acc = tuned_scores[model_name]['Accuracy']
    tuned_f1 = tuned_scores[model_name]['F1']

    print(f"{model_name:<20} {base_acc:<10.4f} {tuned_acc:<10.4f} {base_f1:<10.4f} {tuned_f1:<10.4f}")
```
